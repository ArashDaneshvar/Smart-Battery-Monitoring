```
import tensorflow as tf
from preprocessing import AudioReader
from preprocessing import Spectrogram
from time import time
```

```
audio_reader = AudioReader(tf.int16, 16000)

filename = 'examples/go_0a9f9af7_nohash_1.wav'
audio = audio_reader.get_audio(filename)
```

```
spectrogram_processor = Spectrogram(16000, 0.04, 0.02)
latencies = []

for i in range(100):
    start = time()
    spectrogram = spectrogram_processor.get_spectrogram(audio)
    end = time()
    latencies.append(end - start)
```

```
import numpy as np


min_latency_s = np.min(latencies)
median_latency_s = np.median(latencies)
std_latency_s = np.std(latencies)
max_latency_s = np.max(latencies)
min_latency_ms = min_latency_s * 1000.
median_latency_ms = median_latency_s * 1000.
std_latency_ms = std_latency_s * 1000.
max_latency_ms = max_latency_s * 1000.

print(f'Min Latency: {min_latency_ms:.1f}ms')
print(f'Median Latency: {median_latency_ms:.1f}ms') 
print(f'Std: {std_latency_ms:.1f}ms')
print(f'Max Latency: {max_latency_ms:.1f}ms')
```
