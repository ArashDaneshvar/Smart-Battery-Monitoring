# Deepnote - Getting started

Deepnote is a collaborative cloud platform for ML developers. The development environment is organized in `Projects`, that are collections of Jupyter notebooks sharing the same environment setup and files. In the `ML4IoT 24 LABs` workspace, we have shared and private projects.

## Shared Projects
The `ML4IoT 24 LABs` workspace contains the following shared projects:
- `LABs Material`: (read-only) Notebooks, code, and files with all the reference material.
- `Live Coding`: (read-only) Notebooks created during the labs.
- `TeamXX`: projects associated to a student team, where students belonging to the team can create, edit, and run the code needed to solve the labsâ€™ exercises. 

## Private Projects
Private projects are intendend for personal development. 
For creating a private project, go to the `Main Menu` (menu icon on the top left), `Private Projects`, `Create new`, `New project from template`, `ML4IoT Template`, click on the `Use Template` button.     

## Team Projects
Every student will be assigned to a Team project (e.g., `Team 00`), where she/he can collaborate with her/his teammates. The Team project will appear in the `SHARED WITH ME` tab on the sidebar.

## Run your first Notebook
After cloning the template project, open the notebook `Notebook 1` and click on `Run notebook` button. The expected outcome of a working setup is:

```
Python version: OK
TensorFlow version: OK
```

Note that the first execution can be slow, since you have to wait for the machine configuration.
